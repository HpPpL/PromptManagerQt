## Формирование множеств имен

```python
set_xml, set_png = set(), set()
for file in os.listdir():

  if file[-3:] == "xml":
    print(file, file[:-4])
    set_xml.add(file[:-4])
  else:
    print(file, file.split('.')[0])
    set_png.add(file.split('.')[0])
```

## Формирование множества одиночных элементов и их удаление

```python
set_dif = set_xml ^ set_png
print(len(set_dif))
```

```python
for file in os.listdir():
  if file.split('.')[0] in set_dif:
    print(f"{file = } удален!")
    os.remove(f'/content/drive/MyDrive/all/{file}')
```
